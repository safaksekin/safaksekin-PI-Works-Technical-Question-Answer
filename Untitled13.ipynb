```python
import numpy as np
import pandas as pd
df=pd.read_csv("country_vaccination_stats.csv")
df.head()
df_c=df.copy()
df1=df.notna()
df2=df[df1["daily_vaccinations"]==True]
df.tail()
```

```python
countries=df["country"].value_counts()
```

```python
cnt_list=countries.keys()
cnt_list
```

```python
min_list=[]
for j in range(len(cnt_list)):
    min_list.append(df2[df2["country"]==cnt_list[j]]["daily_vaccinations"].min())
```

```python
df.head()
```

```python
for i in range(len(cnt_list)):
    cpy=df.loc[df["country"]==cnt_list[i]]["daily_vaccinations"]
    cpy.fillna(min_list[i],inplace=True)
    df.loc[df["country"]==cnt_list[i],["daily_vaccinations"]]=cpy
```
